```python
% matplotlib inline
import pandas as pd
import tushare as ts
```

```python
data = ts.get_hist_data('510050', '2017-01-01')
data = data.sort_index()

df = pd.DataFrame()
df['close'] = data['close']
df['change'] = df['close'] - df['close'].shift(1)
```

```python
df['ma5'] = df['close'].rolling(window=5, center=False).mean()
df['ma20'] = df['close'].rolling(window=20, center=False).mean()
df
```

```python
df = df.dropna()
df
```

```python
df['pos'] = 0
df['pos'][df['ma5']>=df['ma20']] = 10000
df['pos'][df['ma5']<df['ma20']] = -10000
df['pos'] = df['pos'].shift(1).fillna(0)
df
```

```python
df['pnl'] = df['change'] * df['pos']

df['fee'] = 0
df['fee'][df['pos']!=df['pos'].shift(1)] = df['close'] * 20000 * 0.0003

df['netpnl'] = df['pnl'] - df['fee']
df
```

```python
df['cumpnl'] = df['netpnl'].cumsum()
df['cumpnl'].plot()
```
